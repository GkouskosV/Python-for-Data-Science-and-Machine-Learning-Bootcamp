## The Data

We will be working with a famous titanic data set for these exercises. Later on in the Machine Learning section of the course, we will revisit this data, and use it to predict survival rates of passengers. For now, we'll just focus on the visualization of the data with seaborn:


```python
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

**Join Plot with fare and age variables, to check if is there any correlation between them.** 

```python
sns.jointplot(x='fare', y='age',data=titanic)
```

**Check the distribution of fare variable.**

```python
sns.distplot(titanic['fare'], kde=False, color='red', bins=30)
```

**Apropos the age, compare the class tickets**

```python
sns.boxplot(x='pclass', y='age',data=titanic, palette='rainbow')
```

```python
sns.swarmplot(x='pclass', y='age', data=titanic, palette='rainbow')
```

**How many females and males were in the ship?**

```python
sns.countplot(x='sex', data=titanic)
```

**Check the correlation of the variables in the dataset**

```python
sns.heatmap(titanic.corr(), cmap='coolwarm')
```

**Apropos the age, check the distibution of males and females**

```python
sns.FacetGrid(data=titanic, col='sex').map(plt.hist, 'age')
```

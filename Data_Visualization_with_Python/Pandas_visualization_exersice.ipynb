```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df3 = pd.read_csv('df3')
%matplotlib inline
```

```python
df3.info()
```

```python
df3.head()
```

**Scatter for correlation between 'a' and 'b' variables**

```python
sns.set_style('ticks')
df3.plot.scatter(x='a',y='b', s=50, color='red', figsize=(12,3), lw=1, edgecolors='black')
```

**Histogram of the 'a' column**

```python
df3['a'].plot.hist(color='blue')
```

**Make visualization with ggplot**

```python
plt.style.use('ggplot')
df3['a'].plot.hist(bins=25, alpha=.6)
```

**Create an area plot of all the columns for just the rows up to 30.**

```python
df3.iloc[:30].plot.area(alpha=0.4)
```

**Notice how the legend in our previous figure overlapped some of actual diagram. Can you figure out how to display the legend outside of the plot as shown below?**

```python
df3.iloc[:30].plot.area(alpha=0.4)
plt.legend(loc='center left', bbox_to_anchor=(1,0.5))
```
